### Transforming State names

```python
df.groupby('State')['Enrollment ID'].count().reset_index()
```

```python
df['State']=df['State'].apply(lambda x: x.split(', ')[-1])
```

```python
df['State'] = df['State'].str.strip().str.capitalize()
```

```python
correct_spellings={
    'Dehradun uk':'Uttarakhand',
    'Uttarkhand':'Uttarakhand',
    'Uttrakhand':'Uttarakhand',
    'U.p':'Uttar pradesh',
    'Uttra pradesh':'Uttar pradesh',
    'Uttarpradesh':'Uttar pradesh',
    'New delhi':'Delhi'
}
df['State']=df['State'].replace(correct_spellings)
```

```python
df.groupby('State')['Enrollment ID'].count().reset_index()
```

```python
df['State'].nunique()
```

### Regular transportation mode

```python
df.groupby('Regular transportation mode')['Enrollment ID'].count().reset_index()
```

```python
df['Regular transportation mode'] = df['Regular transportation mode'].str.strip().str.capitalize()
```

```python
correct_forms={ 
    'Campus bus':'College bus',
    'University vehicle':'College bus',
    'Pg':'Walking',
}
df['Regular transportation mode'] = df['Regular transportation mode'].replace(correct_forms)
```

```python
df.groupby('Regular transportation mode')['Enrollment ID'].count().reset_index()
```
